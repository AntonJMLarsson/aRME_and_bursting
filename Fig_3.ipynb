```python
def obs_minus_e(subset):
    c57_subset = c57_allelic_rpkms.reindex(subset, axis=1)
    cast_subset = cast_allelic_rpkms.reindex(subset, axis=1)
    p_subset = (1-(c57_subset == 0).sum(axis=1)/(c57_subset.shape[1] - c57_subset.isnull().sum(axis=1)))*(1-(cast_subset == 0).sum(axis=1)/(cast_subset.shape[1] - cast_subset.isnull().sum(axis=1)))
    p_subset = p_subset.replace(0,np.nan).dropna()
    c57_subset = c57_subset.reindex(p_subset.index)
    biallelic_len = c57_subset.shape[1] - c57_subset.isnull().sum(axis=1)
    c57_subset = c57_subset.replace(np.nan,0)
    cast_subset = cast_subset.reindex(p_subset.index).replace(np.nan,0)
    df_subset = (c57_subset != 0)&(cast_subset != 0)
    return (df_subset.sum(axis=1)/(biallelic_len*p_subset)),df_subset.sum(axis=1), p_subset, biallelic_len
```

```python
def poisson_mean_ll(subset):
    total_subset = np.rint(total_rpkms.reindex(subset, axis=1))
    return total_subset.apply(lambda gene: np.mean(poisson.logpmf(gene, gene).mean()),axis=1)
```

```python
def add_cells_to_clusters(clu): 
    cells = clusters[clusters['cluster'] == clu].index
    print(len(cells))
    print(np.median(obs_minus_e(cells)[0].replace(0,np.nan).dropna()))
    ome_median = {}
    for p in np.linspace(0,100,21):
        print(p)
        p = int(p)
        n = int(np.rint((p/100)*len(cells)))
        print(n)
        ome_median[p] = [np.median(obs_minus_e(np.concatenate([np.random.choice(cells,len(cells)-n, replace=False),clusters.drop(cells).sample(n).index]))[0].replace(0,np.nan).dropna())]
        for i in range(19):
            ome_median[p].append(np.median(obs_minus_e(np.concatenate([cells,clusters.drop(cells).sample(n).index]))[0].replace(0,np.nan).dropna()))
    return ome_median
```

```python
def transfer_cells_to_clusters(from_clu, to_clu): 
    from_cells = clusters[clusters['cluster'] == from_clu].index
    to_cells = clusters[clusters['cluster'] == to_clu].index
    print(len(to_cells))
    print(np.median(obs_minus_e(to_cells)[0].replace(0,np.nan).dropna()))
    ome_median = {}
    for p in np.linspace(0,100,21):
        print(p)
        p = int(p)
        n = int(np.rint((p/100)*len(from_cells)))
        print(n)
        subset = np.concatenate([to_cells,np.random.choice(from_cells,n, replace=False)])
        ome_median[p] = [np.median(obs_minus_e(subset)[0].replace(0,np.nan).dropna())]
        for i in range(19):
            subset = np.concatenate([to_cells,np.random.choice(from_cells,n, replace=False)])
            ome_median[p].append(np.median(obs_minus_e(subset)[0].replace(0,np.nan).dropna()))
    return ome_median
```

```python
def transfer_cells_to_clusters_poisson_ll(from_clu, to_clu): 
    from_cells = clusters[clusters['cluster'] == from_clu].index
    to_cells = clusters[clusters['cluster'] == to_clu].index
    print(len(to_cells))
    print(np.median(obs_minus_e(to_cells)[0].replace(0,np.nan).dropna()))
    ome_median = {}
    for p in np.linspace(0,100,21):
        print(p)
        p = int(p)
        n = int(np.rint((p/100)*len(from_cells)))
        print(n)
        ome_median[p] = [np.median(poisson_mean_ll(np.concatenate([to_cells,np.random.choice(from_cells,n, replace=False)]))[0].replace(0,np.nan).dropna())]
        for i in range(19):
            ome_median[p].append(np.median(poisson_mean_ll(np.concatenate([to_cells,np.random.choice(from_cells,n, replace=False)]))[0].replace(0,np.nan).dropna()))
    return ome_median
```
